## Шаг 2. Предобработка данных

### Названия столбцов

```python
#приведем названия столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

### Типы данных

```python
#выполним преобразование типа данных в столбце "year_of_release", так как год должен быть целым значением
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
data['user_score'].unique()
```

В столбце "user_score" встречается значение "tbd", что означает "to be determined = будет определено". Это то же самое, что и неизвестное значение, поэтому заменим "tbd" на "nan".

```python
#заменим "tbd" на "nan" в столбце "user_score"
data['user_score'] = data['user_score'].replace('tbd', np.nan)

#выполним преобразование типа данных в столбце "user_score", так как рейтинг - это дробное значение
data['user_score'] = data['user_score'].astype('float64')
```

### Пропуски в данных

- name и genre

```python
#посмотрим на пропуски в столбце "name"
data.loc[data['name'].isna()]
```

Ранее мы определили, что в столбцах "name" и "genre" есть всего лишь по 2 пропуска. Сейчас установили, что эти пропуски совпадают. Получается, что у двух позиций мы не знаем ни название игры, ни жанр. Возможно, произошла какая-то ошибка при сборе данных. Так как пропущено всего лишь 2 значения (0,01%) и данная информация будет дальше использоваться в анализе, а точно заменить пропущенные значения невозможно, удалим эти строки.

```python
#удалим строки, где есть пропуски в столбце "name"
data.dropna(subset=['name'], inplace=True)
```

- year_of_release и rating

Чтобы обработать пропуски в столбцах "year_of_release" и "rating" можно попробовать посмотреть на игры, которые представлены на нескольких платформах, ведь если это одна и та же игра, то год выхода и возрастная категория должны иметь одинаковые значения.

```python
#построим сводную таблицу для игр, которые встречаются на нескольких платформах
games = data.pivot_table(index='name', values='platform', aggfunc='count')
games = games.loc[games['platform'] > 1].sort_values(by='platform', ascending=False)
games
```

```python
#проанализируем игру, которая встречается на наибольшем количестве платформ
data.loc[data['name']=='Need for Speed: Most Wanted']
```

Мы видим, что игра "Need for Speed: Most Wanted" выходила на различных платформах и в разные годы (2005, 2012 и 2013), при этом игре присвоено два разных рейтинга Т и Е в одном и том же году. Получается, у одной игры может быть несколько версий и возрастных категорий, и заполнить пропущенные значения в столбцах "year_of_release" и "rating" по названию игры не получится.

Так как в столбце "year_of_release" пропущено всего лишь около 1% значений - оставим пропуски как есть. В столбце "rating" пропущено уже около 40% значений, поэтому заменим пропуски на "unknown".

```python
#заменим пропущенные значения в столбце "rating" на "unknown"
data['rating'] = data['rating'].fillna('unknown')
```

- critic_score и user_score

В столбцах "critic_score" и "user_score" около 50% и 40% пропусков соответственно, однако заполнять их средними или медианными значениями некорректно, так как это может сместить наши оценки. Поэтому заполним пропуски техническим значением "-1".

```python
#заменим пропущенные значения в столбцах "critic_score" и "user_score" на "-1"
data['critic_score'] = data['critic_score'].fillna(-1)
data['user_score'] = data['user_score'].fillna(-1)
```

### Проверка

```python
#посмотрим долю и количество пропусков
pd.DataFrame({'sum_of_null': data.isna().sum(),
              'share_of_null':data.isna().mean()*100
                 }).style.set_precision(2).text_gradient('coolwarm')
```

**Вывод:** Почти все пропуски обработаны, не заполненными остались всего лишь 1,61% значений.

### Cуммарные продажи во всех регионах

```python
#посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data.head()
```

> **Вывод:** предобработка данных выполнена. Изменены названия столбцов и типы данных. Пропуски были обработаны там, где это возможно, а также были посчитаны суммарные продажи. Данные готовы для дальнейшего анализа.

## Шаг 3. Исследовательский анализ данных

### Сколько игр выпускалось в разные годы?

```python
#построим график на основе сводной таблицы, чтобы посмотреть сколько игр выпускалось в разные годы
data.pivot_table(index='year_of_release', values='total_sales', aggfunc=['count','sum']).droplevel(level=1,
axis=1).rename(columns={'count':'количество выпущенных игр','sum':'количество проданных игр'}).plot(xlabel=('Год выпуска'), 
ylabel=('Количество'), title=('Сколько игр выпускалось в разные годы?'));
```

**Вывод:** на графике видим, что до 1990 количество выпускаемых игр было незначительным, это связано с тем, что технологии не были так сильно развиты. У компьютеров не было мощных видео карт и процессоров, был маленький объем доступной оперативной памяти. Однако, начинания с 90-х бум компьютерных игр стал только набирать обороты, а индустрия расти и расти. В первом десятилетии 21-го века игровая индустрия идет вперед семимильными шагами. Ежегодно выпускаются тысячи игр, расходящиеся по всему миру миллионами копий. Успех продлился где-то до 2010. Скорее всего это связано с тем, что в 2007 году Стив Джобс представил первый iPhone и внимание людей переключилось с компьютерных игр на смартфоны.

### Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков?

```python
#напишем функцию, чтобы построить графики и посчитать корреляцию
def plot_corr(platform_name):
    platform = data.loc[data['platform']==platform_name]
    
    figure, ax = plt.subplots(1, 2, figsize=(15,5))
    
    platform.plot(x='critic_score', y='total_sales',style='o',ylabel='Продажи', xlabel='Оценка критиков', 
    title=('Корреляция между оценками критиков и количеством продаж'), ax=ax[0])
    correlation1 = platform['total_sales'].corr(platform['critic_score']) 
    
    platform.plot(x='user_score', y='total_sales',style='o',ylabel='Продажи', xlabel='Оценка пользователей',
    title=('Корреляция между оценками пользователей и количеством продаж'), ax=ax[1])
    correlation2 = platform['total_sales'].corr(platform['user_score']) 
    
    figure.suptitle(platform_name, fontsize=15)
    
    plt.show()
    print('Корреляция между оценками критиков и количеством продаж:', correlation1)
    print('Корреляция между оценками пользователей и количеством продаж:', correlation2)
```

```python
#построим диаграмму рассеяния и посчитаем корреляцию для платформы "Wii"
plot_corr('Wii')
```

**Вывод:** Корреляции между оценками критиков и количеством продаж, а также между оценками пользователей и количеством продаж на платформе "Wii" слабые, всего лишь 0,2. Это значит, что продажи на этой платформе не зависят ни от оценки критиков, ни от оценки пользователей.

### Соотнесем выводы с продажами игр на других платформах

```python
#создадим переменную, где будут храниться 10 платформ лидеров по продажам в актуальном периоде
platform_top10 = list((data.pivot_table(index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales', 
ascending=False)).head(10).index)
```

```python
#применим функцию для платформ-лидеров, кроме платформы "Wii", которую мы уже проанализировали
for platform in platform_top10:
    if platform != 'Wii':
        plot_corr(platform)
```

```python
#чтобы удобнее было делать выводы, соберем все корреляции в одну таблицу
corr_critic = []
corr_user = []
for plat in platform_top10:
    platform = data.loc[data['platform']==plat]
    
    correlation1 = platform['total_sales'].corr(platform['critic_score'])
    correlation2 = platform['total_sales'].corr(platform['user_score'])
    
    corr_critic.append(correlation1)
    corr_user.append(correlation2)
    
correlation_table = pd.DataFrame({'critic_score':corr_critic,'user_score':corr_user}, index=platform_top10)

#пропишем категорию для корреляции
def categorize_correlation(correlation):
    if correlation <= 0.2:
        return 'очень слабая'
    if correlation <= 0.4:
        return 'слабая'
    if correlation <= 0.5:
        return 'средняя'
correlation_table.insert(1, 'critic_corr_category', correlation_table['critic_score'].apply(categorize_correlation))
correlation_table.insert(3, 'user_corr_category', correlation_table['user_score'].apply(categorize_correlation))

correlation_table
```

**Вывод:** Практически у всех платформ корреляции между оценками критиков и количеством продаж, а также между оценками пользователей и количеством продаж слабые. Это значит, что данные оценки почти не влияют на количество продаж. Однако, можно заметить, что в большинстве случаев первая корреляция выше второй. Это значит, что пользователи больше прислушиваются к экспертному мнению критиков, чем пользователей перед покупкой.

### Самые популярные платформы (топ-5) для пользователя каждого региона (NA, EU, JP)

Составим портрет каждого региона, чтобы спланировать продажи на 2017 год, для этого проанализируем данные с 2013 по 2016 год. Чтобы в анализ не попали уже отжившие тренды. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Также посмотрим какая была ситуация, если анализировать период с 2006 по 2016 год.

```python
#создадим список регионов
region_sales = ['na_sales', 'eu_sales', 'jp_sales']

#напишем функцию, чтобы построить график на основе сводной таблицы для каждого региона
def plot_region(parametr, dataset):
    figure, ax = plt.subplots(1, 3, figsize=(15,5))
    label = {'platform':'Платформа', 'genre':'Жанр', 'rating':'Рейтинг'}
    name = {'na_sales':'Северная Америка', 'eu_sales':'Европа', 'jp_sales':'Япония'}
        
    for region in region_sales:
        dataset.groupby(parametr)[region].agg('sum').sort_values(ascending=False).head(5).plot(
        ax=ax[region_sales.index(region)], title=(name[region]), ylabel=('Продажи'), xlabel=(label[parametr]), kind='bar')

    if dataset is data:
        if parametr == 'platform':
            figure.suptitle('Самые популярные платформы с 2006 по 2016 год', fontsize=15)
        if parametr == 'genre':
            figure.suptitle('Самые популярные жанры с 2006 по 2016 год', fontsize=15)
        if parametr == 'rating':    
            figure.suptitle('Влияние рейтинга ESRB на продажи с 2006 по 2016 год', fontsize=15)
    if dataset is data_forecast:
        if parametr == 'platform':
            figure.suptitle('Самые популярные платформы с 2013 по 2016 год', fontsize=15)
        if parametr == 'genre':
            figure.suptitle('Самые популярные жанры с 2013 по 2016 год', fontsize=15)
        if parametr == 'rating':    
            figure.suptitle('Влияние рейтинга ESRB на продажи с 2013 по 2016 год', fontsize=15)
```

```python
plot_region('platform', data)
```

**Вывод:** в Северной Америке самая популярная платформа "X360", в Европе - "PS3", а в Японии - "DS".

```python
plot_region('platform', data_forecast)
```

**Вывод:** на графике видим топ-5 платформ для запуска продаж для каждого региона. Рынки Северной Америки и Европы очень похожи между собой. В обоих регионах популярны "PS4", "Xone", "3DS" и "PS3". Отличие лишь в том, что в Северной Америке люди предпочитают играть на "X360" вместо "PC". В Японии самая популярная платформа "3DS", также в топ-5 входят "PS4", "PSV", "PS3" и "WiiU".

## Шаг 5. Проверка гипотез

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Для проверки гипотез выберем t-тест для двух независимых выборок (двухвыборочный t-критерий), который проверяет гипотезу о равенстве средних в двух выборках (предполагается нормальность распределения переменных, а также равенство дисперсий выборок), так как значения в двух выборках не составляют пару коррелирующих значений.

Сформулируем нулевую и альтернативную гипотезы:

- Нулевая гипотеза H₀: cредние пользовательские рейтинги платформ Xbox One и PC одинаковые
- Альтернативная гипотеза H₁: cредние пользовательские рейтинги платформ Xbox One и PC разные

```python
# Создадим переменные с соответствующими данными по платформам
xone = data.loc[data['platform']=='XOne']['user_score']
pc = data.loc[data['platform']=='PC']['user_score']

alpha = 0.05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(xone, pc, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:** Значение p-value оказалось меньше порогового значения (альфа), наш результат является статистически значимым. Это говорит о том, что наблюдаемое явление действительно имело место, и нулевую гипотезу нужно отклонить. Получается **средние пользовательские рейтинги платформ Xbox One и PC различаются.**

### Гипотеза 2. Средние пользовательские рейтинги жанров Action и Sports  разные

Сформулируем нулевую и альтернативную гипотезы:

- Нулевая гипотеза H₀: cредние пользовательские рейтинги жанров Action и Sports одинаковые
- Альтернативная гипотеза H₁: cредние пользовательские рейтинги жанров Action и Sports разные

```python
# Создадим переменные с соответствующими данными по платформам
action = data.loc[data['genre']=='Action']['user_score']
sports = data.loc[data['genre']=='Sports']['user_score']

alpha = 0.05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(action, sports, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:** Значение p-value оказалось меньше порогового значения (альфа), наш результат является статистически значимым. Это говорит о том, что наблюдаемое явление действительно имело место, и нулевую гипотезу нужно отклонить. Получается **cредние пользовательские рейтинги жанров Action и Sports различаются.**

> **Вывод:** мы проверили гипотезы о равенстве средних двух генеральных совокупностей. Были получены следующие заключения:
> - средние пользовательские рейтинги платформ Xbox One и PC различаются
> - cредние пользовательские рейтинги жанров Action и Sports различаются
