## Предобработка данных

### Изменение типа данных и заполнение пропущенных значений

- тип жилья

```python
#заполним пропущенные значения в столбце "is_apartment"
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Скорее всего пропуски в типе жилья связаны с тем, что люди, которые заполняли анкету, не могут отличить апартаменты от квартиры. По данным сайта https://www.novostroy-spb.ru, апартаменты в Петербурге занимают относительно небольшую долю рынка, порядка 8% от общего объема жилищного строительства. Поэтому логично будет заменить пропуски на "False".

- балкон

```python
#заполним пропущенные значения в столбце "balcony" и приведем их целочисленному типу
data['balcony'] = data['balcony'].fillna(0).astype(int)
```

Скорее всего пропуски в информации о балконах могли появиться из-за того, что людям, которые заполняли анкету, было нечего в нее вписать. Значит в этих квартирах нет балконов. Нужно заменить пропуски на "0".

- пруды и парки

```python
#заполним пропущенные значения в столбце "parks_around3000" и "ponds_around3000" и приведем их целочисленному типу
data['parks_around3000'] = data['parks_around3000'].fillna(0).astype(int)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0).astype(int)
```

Ранее мы определили, что в данных присутствует почти одинаковое количество пропусков в этих столбцах, скорее всего эти данные взаимосвязаны и пруды находятся только в парках. Cкорее всего, как и с балконами, пропуски означают, что рядом нет прудов и парков. Нужно заменить пропуски на "0".

- количество этажей

```python
#удалим строки, где есть пропуски в столбце "floors_total" и приведем их целочисленному типу
data.dropna(subset=['floors_total'], inplace=True)
data['floors_total'] = data['floors_total'].astype(int)
```

В этом столбце пропуски неслучайные, возможно люди не захотели указать информацию, что квартира находится на последнем этаже.
Так как пропущено всего лишь 86 значений (0,4%) и данная информация будет дальше использоваться в анализе, а точно заменить пропущенные значения невозможно, удалим эти строки.

- название населенного пункта

```python
#удалим строки, где есть пропуски в столбце "locality_name"
data.dropna(subset=['locality_name'], inplace=True)

#посмотрим сколько значений было в столбце "locality_name" до преобразований
old_locality_name = len(data['locality_name'].unique())

#чтобы обработать неявные дубликаты, приведем все к нижнему регистру и заменим буквы "ё" на "е"
data['locality_name']=data['locality_name'].str.lower().str.replace('ё', 'е')

#напишем функцию, которая создаст столбец с типами населенных пунктов
def locality_type(name):
    if pd.isna(name):
        return 'unknown'
    if 'санк' in name:
        return 'город'
    if any(x in name for x in ['городск', 'коттедж', 'котедж']):
        return 'поселок городского типа'
    if 'посел' in name:
        return 'поселок'
    if 'деревн' in name:
        return 'деревня'
    if 'село' in name:
        return 'село'
    if 'сад' in name:
        return 'садовое товарищество'
    else:
        return 'область'
data.insert(loc=15, column='locality_type', value=data['locality_name'].apply(locality_type))
```

```python
#напишем функцию, которая очистит наименования от типа населенного пункта
def cleaning_name_type(column):
    clean_names_column = column.str.replace("|".join(["поселок", "село", "деревня", "городского", "коттеджный",
                                                      "городской", "садоводческое", "некоммерческое",
                                                      "садовое", "товарищество","типа"]), "", regex=True).str.strip()
    return clean_names_column

data.insert(loc=16, column='locality_clean_name', value=cleaning_name_type(data['locality_name']))

#посмотрим сколько значений стало в столбце "locality_name" после преобразований
new_locality_name = len(data['locality_name'].unique())

print(f'Было {old_locality_name} уникальных значений, а стало {new_locality_name}')
```

- высота потолков

```python
#заполним пропущенные значения в столбце "ceiling_height"
for locality_type in data['locality_type'].unique():
    data.loc[(data['locality_type'] == locality_type)&(data['ceiling_height'].isna()),'ceiling_height'] = \
    data.loc[data['locality_type'] == locality_type]['ceiling_height'].median()
    
#напишем функцию, которая обработает явные аномальные значения
def ceiling_fix(ceiling):
    median = data['ceiling_height'].median()
    if ceiling > 10:
        return ceiling / 10
    else:
        return ceiling

data['ceiling_height'] = data['ceiling_height'].apply(ceiling_fix)

data.loc[(data['ceiling_height'] < 1.5) | (data['ceiling_height'] > 6), 'ceiling_height'] = data['ceiling_height'].median()
```

Скорее всего аномальные значения в столбце "ceiling_height" возникли из-за того, что люди вводили значения невнимательно или не в тех единицах измерения, поэтому получилось 27 метров вместо 2.7 метров. Также 12 в данных может быть, напрмер, в футах и на деле означать 3,6 метра.

- кухня и гостинная

```python
#чтобы заполнить столбцы "kitchen_area" и "living_area", рассмотрим данные отдельно по типу квартиры (студия или нет)
#Начнем анализ с площади кухни в студиях:
data.loc[data['studio']==True]['kitchen_area'].unique()
```

Мы видим, что во всех студиях не указана площадь кухни, скорее всего это связано с тем, что в студии нельзя оценить площадь кухни, так как это пространство не выделено в отдельную комнату. Поэтому заменим все пропущенные значения на "0".

```python
#заполним пропущенные значения в столбце "kitchen_area" в студиях
data.loc[(data['studio'] == True)&(data['kitchen_area'].isna()),'kitchen_area'] = 0
```

```python
#расчитаем медианное процентное соотношение жилой зоны к общей площади в студиях:
living_ratio_studio = data.loc[data['studio']==True]['living_area'].median() / \
data.loc[data['studio']==True]['total_area'].median()

#заполним пропущенные значения в столбце "living_area" в студиях
data.loc[(data['studio'] == True)&(data['living_area'].isna()),'living_area'] = \
data.loc[(data['studio'] == True)&(data['living_area'].isna()),'total_area'] * living_ratio_studio
```

```python
#расчитаем медианное процентное соотношение кухни к общей площади в обычных квартирах:
kitchen_ratio_flat = data.loc[data['studio']==False]['kitchen_area'].median() / \
data.loc[data['studio']==False]['total_area'].median()

#расчитаем медианное процентное соотношение жилой зоны к общей площади в обычных квартирах:
living_ratio_flat = data.loc[data['studio']==False]['living_area'].median() / \
data.loc[data['studio']==False]['total_area'].median()

#заполним пропущенные значения в столбце "kitchen_area" в обычных квартирах
data.loc[(data['studio'] == False)&(data['kitchen_area'].isna()),'kitchen_area'] = \
data.loc[(data['studio'] == False)&(data['kitchen_area'].isna()),'total_area'] * kitchen_ratio_flat

#заполним пропущенные значения в столбце "living_area" в обычных квартирах
data.loc[(data['studio'] == False)&(data['living_area'].isna()),'living_area'] = \
data.loc[(data['studio'] == False)&(data['living_area'].isna()),'total_area'] * living_ratio_flat
```

- дата публикации

```python
#изменим тип данных в столбце "first_day_exposition"
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

- сколько дней было размещено объявление

```python
#заполним пропущенные значения в столбце "days_exposition" и приведем их целочисленному типу
data.loc[data['days_exposition'].isna(), 'days_exposition'] = (pd.Timestamp.now() - data['first_day_exposition']).dt.days
data['days_exposition'] = data['days_exposition'].astype(int)
```

Пропуски в этом столбце могут означать то, что объявление еще не закрыто, поэтому мы не можем заполнить их "0". Нужно заменить пропущенные значения на количество дней, между текущей датой и датой размещения объявления.

- цена

```python
#изменим тип данных в столбце "last_price"
data['last_price'] = data['last_price'].astype('int')
```

### Удаление аномалий

Мы частично обработали аномальные значения в потолках. Остаются аномальные значения в общей площади, количестве комнат, цене, площади жилой зоны и кухни. Скорее всего это связано с тем, что это не квартиры, а таунхаусы или дома. 

Так как задача исследования состоит в том, чтобы изучить квартиры, удалим эти строки. Так как количество комнат, общая площадь, цена и т.д. - это взаимосвязанные значения, можно удалить данные лишь в одном столбце, например, "total_area".

```python
#удалим данные в столбце "total_area" с учетом 1.5 размаха
q1 = data['total_area'].quantile(0.25)
q3 = data['total_area'].quantile(0.75)
IQR = q3 - q1
data = data.loc[(data['total_area'] < q3 + 1.5*IQR)&(data['total_area'] > q1 - 1.5*IQR)]
```

```python
new_number_of_rows = data.shape[0] #посчитаем сколько строк стало в таблице
data.describe()
```

**Вывод:** Аномальные значения удалены. Теперь максимальное значение общей площади 114 квадратных метров, комнат - 7, высота потолков - 5,8 метров.

```python
print(f'При этом мы удалили всего лишь: {(old_number_of_rows - new_number_of_rows)/old_number_of_rows :.0%} строк')
```

## Расчеты и добавление в таблицу новых столбцов 

```python
#посчитаем цену одного квадратного метра и вставим этот столбец после "цены"
data.insert(2, 'price_of_one_metr', round((data['last_price'] / data['total_area']).astype(int)))
```

- день недели, месяц и год публикации объявления

```python
#определим в какой день недели было сделано объявление
data['day_of_the_week'] = pd.DatetimeIndex(data['first_day_exposition']).weekday

#определим в какой месяц было сделано объявление
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month

#определим в какой год было сделано объявление
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

- тип этажа квартиры (первый, последний или другой)

```python
#напишем функцию для категоризации этажей
def floor_type(row):
    floor = row['floor']
    if floor == row['floors_total']:
        return 'последний'
    if floor == 1:
        return 'первый'
    else:
        return 'другой'
        
data['floor_category'] = data.apply(floor_type, axis=1)
```

```python
#определим сколько километров до центра 
data['cityCenter_km'] = round(data['cityCenters_nearest'] / 1000)
```

### Какие факторы больше всего влияют на общую (полную) стоимость объекта?

```python
#напишем функцию, чтобы создать сводную таблицу с данными по медианной цене и нужным нам параметрам, построить график,
#а также чтобы рассчитать корреляцию
def plot_corr(column):
    pivot = data.pivot_table(index=(column), values='last_price_mln', aggfunc='median')
    correlation = data[column].corr(data['last_price_mln'])
    return pivot.plot(y = 'last_price_mln', style = 'o', ylabel=('last_price_mln')) and print(f'Корреляция: {correlation}')
```
